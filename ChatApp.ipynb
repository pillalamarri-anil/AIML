```python

from google.colab import userdata
api_key = userdata.get('OPENAI_API_KEY')
end_point = userdata.get('API_END_POINT')

if not api_key:
    raise ValueError(
        "OPENAI_API_KEY not found! "
        "Make sure you have a .env file with: OPENAI_API_KEY=sk-..."
    )

print("API key loaded successfully.")
```

```python
try:
    from azure.openai import AzureOpenAI
except ImportError:
    # Fallback for Python 3.13 where azure-openai might not be available yet
    from openai import AzureOpenAI

client = AzureOpenAI(api_key=api_key,
            api_version='2024-02-01',
            azure_endpoint=end_point)
print("OpenAI client ready.")
```

```python
models = client.models.list()

for m in models.data:
    print(m.id)
```

```python
from pprint import pprint

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are extremely sweet and helpful assistant."},
        {"role": "user", "content": "What is your name?"}
    ]
)



pprint(response.choices[0].message.content)
```

```python
messages = [
    {"role": "system", "content": "You are a python tutor."
    "You explain concepts very well with examples"
    }
]

def pretty_print(*args):
    text = " ".join(str(args) for args in args)
    try:
        print(textwrap.fill(text, width=80))
    except Exception as e:
        print(text)  # fallback to normal print if text is not a string

while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit", "bye"]:
        break

    messages.append({"role": "user", "content": user_input})

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        temperature=0,
        max_tokens=100)

    reply = response.choices[0].message.content
    pretty_print(f"Assistatn: {reply}\n")

    messages.append({"role": "assistant", "content": reply})
```
